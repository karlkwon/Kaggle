## Categorical data 2 number (Dummy variable)

```python
# http://rkzdatamining.com/2015/09/10/dataanalysis_pandas_python/
def CategConvert():
    # read the data using pandas module
    data = pd.read_csv('example.csv')
    # drop the numeric columns
    data2 = data.drop(['numeric'], axis = 1 )
    # convert the categorical values of the column to independent binary features
    data2_dumm = pd.concat([pd.get_dummies(data2[col]) for col in data2], axis=1)
    '''concatenate the binary features with the numeric features by 
    removing the original cateorical column '''
    cols_to_use = data.columns - data2.columns
    dfNew = pd.merge(data[cols_to_use], data2_dumm, left_index=True, right_index=True, how='outer')
    dfNew.to_csv('convertedata.csv')
```

## Make function...

```python
def preProcessing(df, df_min=0, df_max=0):
    isTrain = 'country_destination' in df
    
    if isTrain:
        df_2 = df.drop(['Unnamed: 0', 'date_account_created', 'timestamp_first_active', 'age', 'id', 'country_destination'], axis=1)
    else:
        df_2 = df.drop(['Unnamed: 0', 'date_account_created', 'timestamp_first_active', 'age', 'id'], axis=1)
        
    df_2_ = df[['date_account_created', 'timestamp_first_active', 'age']]
    
    # date to number (sec)
    tmp = df_2_['date_account_created']
    tmp_dac = pd.to_datetime(tmp).astype(np.int) / 1000 / 1000 / 1000
    df_2_['date_account_created'] = tmp_dac

    tmp = df_2_['timestamp_first_active']
    tmp_tfa = pd.to_datetime(tmp, format='%Y%m%d%H%M%S').astype(np.int) / 1000 / 1000 / 1000
    df_2_['timestamp_first_active'] = tmp_tfa

    if isTrain:
        df_min = df_2_.min(axis=0)
        df_max = df_2_.max(axis=0)
    
    # normalize
    df_2_ = (df_2_ - df_min)/(df_max - df_min)
    
    # category value to number (dummy variable)
    tmp_ret = df_2_
    for d in df_2[0:1]:
        tmp = pd.get_dummies(df_2[d])
        cols = [ (d.encode('utf-8') + '_'.encode('utf-8') + str(s).encode('utf-8')).decode('utf-8') for s in tmp.columns.values]
        tmp.columns = cols

        tmp_ret = tmp_ret.join(tmp)
    
    if isTrain:
        y = df['country_destination']
        return tmp_ret, y, df_min, df_max
    else:
        return tmp_ret
```

```python
tmp_train, target, df_min, df_max = preProcessing(df_train)
```
