# Cross validation

```python
from sklearn.cross_validation import KFold

len = df_train.shape[0]
kf = KFold(len, n_folds=5)
kf
```

```python
np_train = df_train.as_matrix()
np_target = df_target.as_matrix()
```

```python
print(np_train[0], np_target[0])
```

```python
from sklearn.metrics import confusion_matrix

labels = np.unique(np_target)

k_ = 5

cf = None

for train_index, test_index in kf:
    X_train, X_test = np_train[train_index], np_train[test_index]
    y_train, y_test = np_target[train_index], np_target[test_index]
    
    clf = neighbors.KNeighborsClassifier(k_, weights='uniform')
    clf.fit(X_train, y_train)
    
    Z = clf.predict(X_test)
    
    y_test = y_test.T.flatten()
    
    print(labels)
    if cf is None:
        cf = confusion_matrix(Z, y_test, labels=labels)
    else:
        cf = cf + confusion_matrix(Z, y_test, labels=labels)
        
print(cf)
```

```python
print(cf.sum(), cf.diagonal().sum())
```

```python
print("precision: ", cf.diagonal().sum()/cf.sum()*100)
```
